## 2. populate quesiton and add BIO labels

```python
form = pd.read_excel('/content/drive/MyDrive/text2elasticsearch/Template-Text2ESQ.xlsx')
```

```python
def populate(value,form):
  d = []
  Q_raw = form.Questions.apply(lambda x: x.split("##"))
  Q_key = [re.findall(r'(\s*\[)(.*)(\])',key[0])[0][1] for key in Q_raw]
  Q_d = {}
  for k,v in zip(Q_key,Q_raw):
    Q_d[str(k)]=v
  for key in value.keys():
    Q = [re.sub(r'\s*\[.*\]',' '+value[key][i],Q_d[key][int(i/100)]) for i in range(100*len(Q_d[key]))]
    for Ques in Q:
      d.append({'Field':key , 'Question':Ques})

  return d
```

```python
def BIO_TAGGING(form,value):
  d = []
  Q_raw = form.Questions.apply(lambda x: x.split("##"))
  Q_key = [re.findall(r'(\s*\[)(.*)(\])',key[0])[0][1] for key in Q_raw]
  Q_d = {}
  Q_tag = {}
  for k,v in zip(Q_key,Q_raw):
    Q_d[str(k)]=v
  for key in value.keys():
    tag = []
    for i in range(len(Q_d[key])):
      s = re.sub(r'[\[\]]','0',Q_d[key][i])
      a = re.sub(r'\b(?!0\w+0)\w+',' o ',s)
      tag.append(re.sub('[\.\?]',' o ',a))
      Q_tag[key]= tag
  tag_dic = {}
  for key in value.keys():
    tag_list=list(range(len(value[key])))
    for i in range(len(value[key])):
      if value[key][i] == nltk.word_tokenize(value[key][i]):
        tag_list[i]= 'B-'+key
      else:
        token = nltk.word_tokenize(value[key][i])
        for j in range(len(token)):
          if j == 0:
            tag_list[i] = 'B-'+key+' '
          else:
            tag_list[i] += 'I-'+key+' '
    tag_dic[key] = tag_list
    Q_BIO = [re.sub(r'\s*0.*0',' '+tag_dic[key][i],Q_tag[key][int(i/100)]) for i in range(100*len(Q_tag[key]))]
    for Ques in Q_BIO:
      d.append({'Field':key , 'tagging':Ques})

  return d
```

```python
key = 'ALLERGIES'
for i in range(len(values[key])):
  if nltk.word_tokenize(values[key][i]) == ''.join(str.split()):
    tag_list[i]= 'B-'+key
  else:
    token = nltk.word_tokenize(values[key][i])
    for j in range(len(token)):
      if j == 0:
        tag_list[i] = 'B-'+key+' '
      else:
        tag_list[i] += 'I-'+key+' '
     
```
